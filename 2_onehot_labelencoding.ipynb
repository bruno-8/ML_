**Task 2:** adults.csv dataset 

```python
df = pd.read_csv('adult.csv')
df
```

```python
df['relationship'].unique()
```

```python
# To encode categorical data into numeric, we use Label encoding;
# ML Algos only work on numerical data
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['relationship_coded'] = label_encoder.fit_transform(df['relationship'])
#existing relationship values will be converted into numerical data and stored in a new column called relationship_coded
```

```python
df.head(20)
```

```python
# To overcome the problem of orders in Label Encoding, we apply One Hot encoding
# the values in the column on which we are applying one hot encoding, become new separate columns with 0 or 1 as their values,
# 0 signfies absence and 1 signifies presence of that particular value
```

```python
df = pd.get_dummies(df, columns = ['relationship'])
```
