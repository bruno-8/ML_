```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#1. generating a scatter plot
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#plotting the above scatter plot, first parameter is the x-axis feature and second parameter is the y-axis feature
# c='r' denotes the colour of the data values plotted
plt.scatter(x,y, c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot graph in 2D')
plt.savefig('scatterplt.png')
```

```python
#2. create a subplot with diff markers and line colors
#the third parameter in the .plot function signifies the color and type of line plotted for the graph
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
#plot a bar plot for the given x and y values
#bar plots show the count for any variable. They are univariate graphs
x=[2,8,10]
y=[11,16,9]
plt.bar(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Graph')
plt.show()
```

```python
#Plot a box plot for the given values and state your inferences
a=[3,4,5,7,9,8,12,13,7,8,19,90,12,15]
plt.boxplot(a)
```

The inferences from the above box plot are as follows:
min= 3;
max=19;
outlier is 90 and the plot is right-skewed

```python
#Read the file into the program
df = pd.read_csv('seeds.csv')
```

```python
#explore the head of the dataset
df.head()
```

```python
df['Type'].unique()
```

```python
# understanding the count of seeds of each type
df['Type'].value_counts(ascending = True)
```

```python
#plotting a bar plot for the count obtained
df["Type"].value_counts().plot.bar()
```

```python
#plot a scatter plot for the given count
sns.countplot('Type', data = df)
```

```python
#plot a scatter plot for kernel width vs length
plt.scatter(df['Kernel.Length'], df['Kernel.Width'], c = 'r')
```

Inferences:
Min Length is 4 and the corresponding width is 2.7
Max Length is 66.6 and the corresponding width is 4.2

```python
sns.scatterplot(x='Kernel.Length', y = 'Kernel.Width', data = df)
```

```python
#to plot and give a hue to the data values according to the column name "Type"
sns.scatterplot(x='Kernel.Length', y = 'Kernel.Width', hue = 'Type' , data = df)
```

This graph helps us to better understand the data in our dataset by grouping and giving a colour to the data values according to their "Type".

```python
#palette gives us specific shades of colours which are used in the graph
sns.scatterplot(x='Kernel.Length', y = 'Kernel.Width', hue = 'Type', data = df, palette = 'deep')
```

```python
#plotting a joint plot i.e. a scatter plot combined with a histogram
# scatter plot in the middle, a histogram at the top for the x-axis variable and a histogram on the left for the y-axis variable
sns.jointplot(x='Perimeter', y = 'Compactness', data = df)
```

```python
#Box plot to understand the relation between type and compactness
sns.boxplot(x = 'Type', y='Compactness', data=df)
```

Inference:
* Type 1 and Type 2 have similar Minimum Compactness.
* Median of these two types is almost the same.
* No outliers are present.
* Lot of the Compactness of Type 3 is at the lower range.

```python
#Construct a pair plot. Since there are 8 columns, the pair plot will be displayed within an 8x8 grid
#Every column plots a graph with every other column in the dataset
# hue ='Type' is used since it is mentioned that Type is the main parameter. 
# diag_kind = 'hist' denotes the kind of graph that will be plotted across the diagonal of the grid
sns.pairplot(df, hue = 'Type', diag_kind = 'hist' )
```

**Observations:**
* There is a positive relationship b/w Perimeter and Area.
* Asymmetry.Coeff has no relationship with any other variables.
* Type 2 seeds are at a higher level while Type 3 seeds are at a lower level.
* Since Asymmetry.Coeff has no relationships with variables, we can drop it from the pair plot.

```python
#Plot a Violin Plot
# In a violin plot, vertically we have a box plot and horizontally we have histograms on either side
#in the narrow part, the lower freq value components are present
#the wider part contains the higher freq value components are present.
sns.violinplot(x = 'Type', y = 'Compactness', data = df)
```

Observations:
* Median of Type 3 is lower than the other two types. 

```python
#kernel density exploration graph will show us how the values of variables are distributed in our dataset
sns.FacetGrid(df, hue = 'Type', height = 5).map(sns.kdeplot, 'Compactness').add_legend()
```

Observations:
* Type 1 and 2 are overlapping as compared to Type 3.
* Thus, it is difficult for ML to predict the differences between them.

```python
# pair plot with KDE as the diagonal
sns.pairplot(df, hue = 'Type',  height = 3)
```

Observations:
* PDF of Compactness shows that Type 1 and 2 seeds have similar features in Compactness.
* Thus it is difficult to distinguish between them.
* Based on Kernel.Length, Type 2 and Type 3 are overlapping
* PDF of Asymmetric Coefficient shows excessive overlapping.


```python
# Andrews Curve is used to check separability of the data.
pd.plotting.andrews_curves(df, class_column = 'Type')
```
