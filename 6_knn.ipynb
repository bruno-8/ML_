```python
dataset=[[-1,1,'N'],[0,1,'P'],[0,2,'N'],[1,-1,'N'],[1,0,'P'],[1,2,'P'],[2,2,'N'],[2,3,'P']]
```

```python
dataset
```

```python
import math
from math import sqrt
# Euclidean distance
def ED(row1, row2): #row 1 is testing point and row2 is dataset
    dist = 0
    for i in range(len(row1)-1): #we created a loop in range of the length of our testing data(row1) , just excluding the last column as it is our class
        dist += (row1[i] - row2[i])**2            #(x1-x2)**2+(y1-y2)**2
    return sqrt(dist)
```

```python
def neighbour(train, test_row, num): #1. train data you have 5 Data Points . 2. in test_row you have only 1 point . 3. num we have number of Neighbors
    dist = list() # []
    data = []
    for i in train:
        #dist = ED(test_row, i)
        dist.append(ED(test_row, i))
        data.append(i)

    dist = np.array(dist)
    data = np.array(data)
    index_dist = dist.argsort()
    data=data[index_dist]
    neighbors = data[:num]
    return neighbors
```

```python
def predict_classification(train, test_row, num):
    Neighbors = neighbour(train, test_row, num)
    Classes = []
    for i in Neighbors:
        Classes.append(i[-1])
    prediction = max(Classes, key= Classes.count)
    return prediction
```

```python
test=[[1,1,'']]
prediction = predict_classification(dataset,test,3)
print("k=3 we got:{}".format(prediction))
prediction = predict_classification(dataset,test,5)
print("k=5 we got:{}".format(prediction))
prediction = predict_classification(dataset,test,7)
print("k=7 we got:{}".format(prediction))
```
